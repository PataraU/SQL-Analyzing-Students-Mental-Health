```python
##Explore and analyze the students data to see how the length of stay (stay) impacts the average mental health diagnostic 
scores of the international students present in the study.

Field Name	Description
inter_dom	Types of students (international or domestic)
japanese_cate	Japanese language proficiency
english_cate	English language proficiency
academic	Current academic level (undergraduate or graduate)
age	Current age of student
stay	Current length of stay in years
todep	Total score of depression (PHQ-9 test)
tosc	Total score of social connectedness (SCS test)
toas	Total score of acculturative stress (ASISS test)
```

```python
%%sql

SELECT stay, 
    COUNT(*) AS count_int,
    ROUND(AVG(todep), 2) AS average_phq, 
    ROUND(AVG(tosc), 2) AS average_scs, 
    ROUND(AVG(toas), 2) AS average_as,
    (ROUND(AVG(todep), 2) + ROUND(AVG(tosc), 2) + ROUND(AVG(toas), 2)) AS total_average_score
FROM students
WHERE inter_dom = 'Inter'
GROUP BY stay
ORDER BY stay DESC;
     
```

##Analysis

Depression (average_phq):
Depression scores fluctuate across different lengths of stay. The highest depression score is 13 for a 10-year stay, while the lowest is 0 for a 5-year stay.
Individuals with shorter stays (1 year) have a moderately low depression score (7.48), while those with a stay of 4 to 6 years seem to experience milder depression.

Social Connectedness (average_scs):
Social connectedness generally increases as the length of stay decreases. For example, individuals with a 1-year stay report an average score of 38.11, while those staying for 10 years report a lower score of 32.0.
There is a notable peak in social connectedness for individuals staying 7 and 8 years, with scores of 48.0 and 44.0, respectively.

Acculturative Stress (average_as):
Acculturative stress increases as the length of stay increases, peaking at 5 years (91.0). This could suggest that stress related to cultural adaptation builds up over time and is highest after a mid-length stay.
For shorter stays (1 to 3 years), acculturative stress levels are slightly lower, around the mid-70s range.
Total Score (Total_score):

The total score, which is the sum of the three measures (PHQ-9, SCS, and ASISS), generally increases with the length of stay, peaking at 130.21 for individuals staying 4 years and remaining high for those with 3- to 5-year stays.
The highest total score is seen in the 4- to 5-year range, indicating that individuals staying for this period experience a combination of depression, social connectedness, and acculturative stress at higher levels compared to others.

##Summary

According to the research, people tend to become less socially engaged and gradually experience more acculturative stress 
the longer they stay, with a notable peak occurring around the 5-year mark. Depression scores vary, declining after eight 
years and then increasing once more as the stay shortens to three years or less. Shorter stays—a year, 
for example—are associated with stronger levels of social connectivity and lower levels of acculturative stress. 
This suggests that longer stays may provide more difficulties for social bonds and environment adaptation.
